```python
from matplotlib.lines import lineStyles
import matplotlib.pyplot as plt

demo = data.groupby("year")["amount"].sum()
plt.figure(figsize=(10,6))
plt.plot(demo.index, demo.values,marker="o",markerfacecolor="red",color="black",markeredgecolor="green",linestyle="--",linewidth=1.5,label="Total amount")
plt.title("Year wise Amount investment", color="red",fontsize=16)
plt.grid(linestyle="--",color="skyblue",linewidth=1.5)
plt.xlabel("Year", color = "blue",fontsize=12)
plt.ylabel("Amount",color="blue",fontsize=12)
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt


demo = data["city"].value_counts().head(10)
color = ["blue","red","black","yellow","pink", "purple", "orange", "brown", "green", "skyblue"]

plt.figure(figsize=(10,6))
plt.bar(demo.index, demo.values,color=color,alpha=0.4,edgecolor="black")

plt.xticks(rotation = 45)
plt.title("Top 10 startup company's in India",color="red",fontsize=20)
plt.xlabel("City Names",color = "blue",fontsize=12)
plt.ylabel("Number of startup",color = "red",fontsize=12)
plt.grid(True,color="black")
```

```python
demo = data["vertical"].value_counts().head(10)
color=["blue", "black", "red", "pink", "green", "purple", "brown", "skyblue", "orange", "lightgreen"]

plt.figure(figsize=(10,6))
plt.barh(demo.index[::-1], demo.values[::-1], color=color,alpha=0.6)
plt.xticks(rotation = 45)
plt.title("Top 10 Sectors startup in India",color="red",fontsize=20)
plt.xlabel("Number of startup copmany",color = "blue",fontsize=12)
plt.ylabel("Sectors",color = "red",fontsize=12)
plt.grid(axis = 'x',linestyle="--", color="black")
```

```python
demo = data["round"].value_counts().head(7)
color=["lightgreen", "skyblue", "lightblue", "pink", "green", "purple", "brown"]

plt.figure(figsize=(10,6))
plt.pie(demo.values, labels=demo.index,autopct="%1.1f%%",colors=color)
plt.xticks(rotation = 45)
plt.title("Founding round ditrubution",color="red",fontsize=20)


```

```python
# 5. How are startup funding amount distributed?

demo = data["amount"].value_counts().head(10)
color=["lightgreen", "skyblue", "lightblue", "pink", "green", "purple", "brown"]
plt.figure(figsize=(10,6))
plt.hist(data["amount"],bins=10,edgecolor="black")
plt.title("Distributed of startup funding",color="blue",fontsize=15)
plt.show()

```

```python
# 6 The Relationship between funding year and Funding Amount?

demo = data["year"].value_counts().head(10)
plt.figure(figsize=(10,6))

plt.scatter(data["year"],data["amount"],color="brown")
plt.yticks()
plt.title("Funding amount by year",color="blue",fontsize=15)
plt.xlabel("year")
plt.ylabel("Amount")
plt.grid(linestyle="--")
plt.show()



```

```python
# 7 which funding round occcurs most frequently?
import seaborn as sns

demo = data["round"].value_counts().head(10).index
plt.figure(figsize=(10,6))
sns.countplot(data, y= "round",order=demo,palette="viridis")
plt.title("Most frequently founding round")
plt.show()

```
